```python
import pyreadstat
import pandas as pd

df, meta = pyreadstat.read_sav("C:\\Users\\Aliprandi\\Desktop\\projectWahlen\\sample_data\\ZA7733_sA_v1-0-0.sav")
```

```python
df
```

```python
df = df.drop(['study', 'sample', 'lfdn'], axis=1)


correlation_matrix = df.corr(numeric_only=True)


print(correlation_matrix)
```

```python
import numpy as np 
```

```python
threshold = 0.95 

corr_mask = np.abs(correlation_matrix) >= threshold
np.fill_diagonal(corr_mask.values, False)

strongly_related_columns = corr_mask.columns[corr_mask.any(axis=1)].tolist()
print(strongly_related_columns)
```

```python
df_filtered = df[strongly_related_columns]

filtered_corr_matrix = df_filtered.corr(numeric_only=True)

distance_matrix = 1 - np.abs(filtered_corr_matrix)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
import matplotlib.pyplot as plt

condensed_dist = squareform(distance_matrix.values, checks=False)

linked = linkage(condensed_dist, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linked, labels=distance_matrix.columns, orientation='top')
plt.title("Hierarchical Clustering of Columns")
plt.show()
```

```python
num_clusters = 2
clusters = fcluster(linked, num_clusters, criterion='maxclust')

cluster_df = pd.DataFrame({'column': distance_matrix.columns, 'cluster': clusters})
print(cluster_df.sort_values('cluster'))
```

```python
for cluster_id in cluster_df['cluster'].unique():
    cols = cluster_df[cluster_df['cluster'] == cluster_id]['column'].tolist()
    print(f"\nCluster {cluster_id} columns: {cols}")
```
